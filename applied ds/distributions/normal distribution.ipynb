# Normal Distribution 

in this the number having the highest probablity of occuring is ZERO <br>
Mean of this curve is zero 
<br>
# Parameters
----------
loc : <br>float or array_like of floats
    Mean ("centre") of the distribution.
    <br><br>
scale : <br>float or array_like of floats
    Standard deviation (spread or "width") of the distribution. Must be
    non-negative.<br><br>
    
size : <br>int or tuple of ints, optional
    Output shape.  If the given shape is, e.g., ``(m, n, k)``, then
    ``m * n * k`` samples are drawn.  If size is ``None`` (default),
    a single value is returned if ``loc`` and ``scale`` are both scalars.
    Otherwise, ``np.broadcast(loc, scale).size`` samples are drawn.
<br>
# Returns
-------
out :<br> ndarray or scalar
    Drawn samples from the parameterized normal distribution.

# By default the number of bins set is 10 so as the sample size increses the thickness of the plot increses too

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,sharex=True)
ax=[ax1,ax2,ax3,ax4]

for n in range(len(ax)):
    ss=10**(n+1)
    sample=np.random.normal(loc=0,scale=1,size=ss)
    ax[n].hist(sample)
```

# now performin same funtion with bins=100 

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,sharex=True)
ax=[ax1,ax2,ax3,ax4]

for n in range(len(ax)):
    ss=10**(n+1)
    sample=np.random.normal(loc=0,scale=1,size=ss)
    ax[n].hist(sample,bins=100)
    ax[n].set_title("n={}".format(ss))
```

```python
%matplotlib widget
y=np.random.normal(loc=0,scale=1,size=10000)
x=np.random.random(size=10000)
plt.figure()
plt.scatter(x,y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec=gridspec.GridSpec(3,3)

top_histogram=plt.subplot(gspec[0,1:])
side_histogram=plt.subplot(gspec[1:,0])
lower_right=plt.subplot(gspec[1:,1:])
                          
```

```python
lower_right.scatter(x,y)
top_histogram.hist(x,bins=100)
s=side_histogram.hist(y,bins=100,orientation="horizontal")
```
