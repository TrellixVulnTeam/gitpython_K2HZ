```python

# import the pandas library
import pandas as pd
df = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
cf = pd.DataFrame(
   {'id':[1,9,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(df)
print(cf)
```

```python
pd.merge(df,cf,on=["subject_id","id"])
```

```python
pd.merge(df,cf,on=["id"])
```

```python
pd.merge(df,cf,left_index=True,right_index=True,how="inner")
```

```python
pd.merge(df,cf,on=["subject_id","id"],how="left")
```

```python
pd.merge(df,cf,on=["subject_id"])
```
