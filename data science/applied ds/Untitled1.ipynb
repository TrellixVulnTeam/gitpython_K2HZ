```python
import pandas as pd
f=open("university_towns.txt","r")
d=f.readlines()
state=""
region=""
df=pd.DataFrame(columns=["State", "RegionName"])
for i in d:
    i=i.strip()
    if i[-6:]=="[edit]":
        state=i[:-6]
        continue;
    region=i
    s=pd.Series({"State":state,"RegionName":region})
    df=df.append(s,ignore_index=True)
df.RegionName=df.RegionName.str.replace("\s\(.*","")
```
